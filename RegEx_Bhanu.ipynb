## {n} Repetition Factor
>  **The {n} syntax specifies that the preceding character (or group) must appear exactly n times.**

```python
pattern = 'a{3}'
test_string = 'aaaaa'
# This will find one match: 'aaa'
matches = re.findall(pattern, test_string)
print(matches)
```

# {n,}
> **The {n,} syntax indicates that the preceding character (or group) must appear at least n times, but it can appear more than that, including infinitely many times.**

```python
import re

pattern = 'a{2,}'
test_string = 'aaa b aa c a'
# This will find two matches: 'aaa' and 'aa'
matches = re.findall(pattern, test_string)
print(matches)  
```

## {n,m} 
> **The {n,m} syntax specifies that the preceding character (or group) must appear at least n times but no more than m times.**

```python
import re

pattern = 'a{2,4}'
test_string = 'a aa aaa aaaa aaaaa'
# This will find matches for 'aa', 'aaa', and 'aaaa'
matches = re.findall(pattern, test_string)
print(matches) 
```

## Parentheses () - Grouping and Capturing
- > **Purpose:** Parentheses are used for two main reasons: to group parts of a pattern together and to capture text matched by that part of the pattern.

- > **Grouping:** Allows you to apply quantifiers to entire patterns, combine multiple expressions, or isolate parts of a regex for logical operations.
- > **Capturing**: Text matched by the portion of the pattern inside parentheses can be captured and later referred to, either for extracting data or for back-references within the pattern.

## Lookhead Lookbehind
- > **lookaheads and lookbehinds are powerful tools that allow you to check for the presence or absence of a pattern without including it in the match results. They're often used when you want to assert conditions on the strings to find matches based on surrounding text.**

- > Lookaheads come in two forms: positive (?=) and negative (?!).

- > **Positive Lookahead (?=...)**: Asserts that the given pattern is ahead. For instance, X(?=Y) will match 'X' only if 'X' is followed by 'Y'.
- > **Negative Lookahead (?!...)**: Asserts that the given pattern is not ahead. For instance, X(?!Y) will match 'X' only if 'X' is not followed by 'Y'.

```python
# positive lookhead 
import re
# Match digits only if they are followed by "th"
text2 = "1st 2nd 3rd 4th 5th"
matches2 = re.findall(r"\d+(?=th)", text2)
print("Example 2:", matches2)
```

```python
#Negative lookhead
# Example 4: Match digits only if they are not followed by "th"
text4 = "1st 2nd 3rd 4th 5th"
matches4 = re.findall(r"\d+(?!th)", text4)
print("Example 4:", matches4)
```

```python
text = "startword something word somethingstart"
pattern = re.compile("(?<=start)word")
matches = pattern.findall(text)

print(matches)  
```

# Flags
> In Pythons re module, regex patterns can be enhanced with various flags. These flags modify the behavior of the regex functions, making them more versatile and adapted to different scenarios

# re.IGNORECASE (or re.I)
> Makes the regex operation **case-insensitive.** This means that characters will match regardless of whether they are in uppercase or lowercase.

```python
import re

text = "Hello World"
pattern = "hello"
matches = re.findall(pattern, text, re.I)
print(matches)  # Output: ['Hello']
```

## re.VERBOSE (or re.X)
> Allows you to write regular expressions that are more readable and self-documented by permitting whitespace and comments within the pattern string. This is useful for complex regex patterns that benefit from additional explanation directly in their context.

```python
pattern = re.compile(r"""
    ^                # beginning of the string
    \d{3}            # area code (3 digits)
    -                # literal dash
    \d{3}            # first 3 digits
    -                # another literal dash
    \d{4}            # last 4 digits
    $                # end of the string
""", re.VERBOSE)

text = "123-456-7890"
match = pattern.search(text)
if match:
    print("Phone number is valid.")
```

## re.MULTILINE (or re.M)
> Changes the behavior of ^ (start of string) and \$ (end of string) anchors. With this flag, \^ matches the start of any line within the string, and $ matches the end of any line, not just the start and end of the entire string.

```python
text = "Starts here\nanother line\nends here"
matches = re.findall(r'^\w+', text, re.MULTILINE)
print(matches)  # Output: ['Starts', 'another', 'ends']
```

# re.DOTALL (or re.S)
> Makes the dot (.) special character match any character at all, including a newline, which normally it would not match. This is useful when you want to match over multiple lines

```python
text = "Hello world.\nHow are you?"
match = re.search('.*', text, re.DOTALL)
print(match.group())  # Output: 'Hello world.\nHow are you?'
```

# re.ASCII (or re.A)
> Makes \w, \W, \b, \B, \d, \D, \s and \S perform ASCII-only matching instead of full Unicode matching. This is useful if you want to ensure that your pattern only matches characters which are represented in the ASCII character set.

```python
text = "Caf"
# Without re.ASCII, the  is considered a word character
matches = re.findall(r'\w+', text)
print(matches)  # Output: ['Caf']

# With re.ASCII, the  is not considered a word character
matches = re.findall(r'\w+', text, re.ASCII)
print(matches)  # Output: ['Caf']
```

```python
user_email = input("Enter your email_id :")

pattern = re.findall("[^0-9\W][\w.]+@(?:gmail|yahoo|outlook)\.(?:com|in|org)",user_email)

if pattern:
    if user_email == pattern[0]:
        print("Valid emailID")
    else:
        print("Invalid EmailID")
else:
        print("Invalid EmailID")
```
